## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
#Create a column that calculates the subtotal for each line using the unit_price and the qty. (6 points)
df['line_subtotal'] = df['unit_price'] * df['qty']
df[['unit_price', 'qty', 'line_subtotal']].head()
```

```python
# Create a column for total weight.
df['total_weight'] = df['unit_weight'] * df['qty']

# Create a column for order weight.
df['t_order_weight'] = df['total_weight']

# Define a function to calculate shipping rate based on order weight.
def shipping_rate(t_order_weight):
    if t_order_weight > 50:
        return t_order_weight * 7
    else:
        return t_order_weight * 10

# Apply the shipping_rate function to the 'order_weight' column and create a new column 'final_shipping'.
df['shipping_rate'] = df['t_order_weight'].apply(shipping_rate)

# Create a column for shipping rate (corrected).
df['final_shipping'] = df['t_order_weight'].apply(shipping_rate)

# Display selected columns for the first 3 rows.
selected_columns = ['unit_price', 'unit_weight', 'qty', 'total_weight', 'final_shipping', 'shipping_rate']
df[selected_columns].head(3)
```

```python
# along with a sales tax of 9.25%
# representing 9.25% as a decimal
sales_tax_rate = 0.0925

# using the subtotal and the shipping price
df['total_price'] = (df['line_subtotal'] + df['shipping_rate']) * (1 + sales_tax_rate)

# Round the 'total_price' column to two decimal places
df['total_price'] = df['total_price'].round(2)

# Display selected columns for the first 3 rows
df[['line_subtotal', 'shipping_rate', 'total_price']].head(3)
```

```python
# Create a column for the cost of each line using unit cost, qty, and shipping rate
df['line_cost'] = df['unit_cost'] * df['qty'] + df['shipping_rate']

# Display selected columns for the first 3 rows
selected_columns = ['line_cost', 'unit_cost', 'qty', 'shipping_rate', 'total_price']
df[selected_columns].head(3)
```

```python
# Create a column for the profit of each line using line cost and line price
df['line_profit'] = df['total_price'] - df['line_cost']
df[['line_profit', 'total_price', 'line_cost']].head(3)
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients, by quantity, spend?
# Check your work from Part 1 for client ids.
top5_client_ids = df.groupby('client_id')['total_price'].sum().nlargest(5).index

def find_total_price(client_id):
    return df.loc[df['client_id'] == client_id, 'total_price'].sum()

for client_id in top5_client_ids:
    total_spent = find_total_price(client_id)
    print(f'{client_id}: ${total_spent:,.2f}')
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. Sort by total profit.
client_summary = df.groupby('client_id').agg({'qty': 'sum', 'shipping_rate': 'sum', 'total_price': 'sum', 'unit_cost': 'sum'})

client_summary['total_profit'] = client_summary['total_price'] - client_summary['unit_cost']

client_summary = client_summary.sort_values(by='total_profit', ascending=False)

client_summary['total_profit'] = client_summary['total_profit'].map('${:,.2f}'.format)

top5_client_summary = client_summary.head(5)

print(top5_client_summary)
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. Sort by total profit.
client_summary = df.groupby('client_id').agg({'qty': 'sum', 'shipping_rate': 'sum','total_price': 'sum', 'unit_cost': 'sum'})

# Sort by total profit.
client_summary['total_profit'] = client_summary['total_price'] - client_summary['unit_cost']
client_summary['total_profit'] = client_summary['total_profit'].map('${:,.2f}'.format)
client_summary['unit_cost'] = client_summary['unit_cost'].map('${:,.2f}'.format)
client_summary['total_price'] = client_summary['total_price'].map('${:,.2f}'.format)
client_summary['shipping_rate'] = client_summary['shipping_rate'].map('${:,.2f}'.format)

client_summary = client_summary.sort_values(by='total_profit', ascending=False)
top5_client_summary = client_summary.head(5)
print(top5_client_summary)
```

```python
# Format the data and rename the columns to names suitable for presentation.

client_summary = client_summary.rename(columns={'qty': 'Total Units Purchased', 'shipping_rate': 'Total Shipping Cost',
                                                'total_price': 'Total Revenue', 'unit_cost': 'Total Cost', 
                                                'total_profit': 'Total Profit'})

top5_client_summary = client_summary.head(5)
top5_client_summary.head(5)
```
